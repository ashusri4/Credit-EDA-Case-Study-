We observe that there are columns having about  48%, 13%, 19% etc. missing values. When dealing with these columns, we  have two simple choices - either delete or retain the column. If we retain the column, we'll have to treat (i.e. delete or impute) the rows having missing values.

If we delete the missing rows, we lose data. If we impute, we introduce bias in our analysis.

Apart from the number of missing values, the decision of either deleting or retaining a variable depends on various other factors as well , such as:

The analysis part which we have , The use of the variable (based on our understanding of the problem), The total size of available data (if we have enough, we can afford to throw away some of it) etc. Thus, for this exercise, we remove the columns having more than missing values and which are not necessary for our analysis.

9) CORELATION MATRIX

```python
#correlation matrix for all numerical columns

corr=target_0.corr()
corr
```

Now, to get rid of the repeated corelated values we are performing following tasks 

```python
#Convert the diagonal and below diagonal values of matrix to False, and replace False with NaN on execution


corr=corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
corr
```

```python
#convert it to dataframe


corrdf = corr.unstack().reset_index()
corrdf.head()
```

```python
#Changing the names of columns


corrdf.columns=['VAR1','VAR2','Correlation']
corrdf.head()
```

```python
#Drop the columns having the missing data

corrdf.dropna(subset = ['Correlation'], inplace = True)
corrdf.head()
```

```python
#Rounding off the values


corrdf['Correlation'] = round(corrdf['Correlation'], 2)
corrdf.head()
```

```python
# Since we see correlation as an absolute value, we are converting it into absolute value

corrdf['Correlation'] = corrdf['Correlation'].abs()
corrdf.head()
```

```python
#Sorting the correlation values

corrdf.sort_values(by = 'Correlation', ascending = False).head(10)
```

We can see that for Target_0 dataframe, Social circle for 30 daysis most correlated followed by  Amt_Goods price and  Amt_credit which are second in the list of  correlation . While, Amt_Credit and Amt_income_total are the least corelated with each other.

```python
#For target_1 dataframe we perform the same operations to find correlation


corr = target_1.corr()

corr = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
corrdf = corr.unstack().reset_index()
corrdf.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf.dropna(subset = ['Correlation'], inplace = True)
corrdf['Correlation'] = round(corrdf['Correlation'], 2)


# Since we see correlation as an absolute value,
# we are converting it into absolute value 
# corrdf['Correlation_abs'] = corrdf0['Correlation'].abs()


corrdf.sort_values(by = 'Correlation', ascending = False).head(10)
```

From the observation above we can say that for target_1 dataframe we can say that Social circle for 30 days and Scoial circle for 60 days has hifhest corelation value followed by Amt_goods_price and Amt_credit  for different days. Thus, the variables correlated in target_0 dataframe and target_1 dataframe are same with slightly varying correlation values.
